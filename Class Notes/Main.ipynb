# Curve Plotting:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to plot
def f(x):
    return x**2 - 4  # Example function: f(x) = x^2 - 4

# Create an array of x values
x = np.linspace(-10, 10, 400)  # 400 points between -10 and 10

# Calculate the corresponding y values
y = f(x)

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(x, y, label='f(x) = xÂ² - 4', color='blue')  # Plot the curve
plt.axhline(0, color='black', lw=0.8, ls='--')  # Add x-axis
plt.axvline(0, color='black', lw=0.8, ls='--')  # Add y-axis
plt.title('Plot of the Function f(x)')  # Title of the plot
plt.xlabel('x')  # Label for x-axis
plt.ylabel('f(x)')  # Label for y-axis
plt.grid()  # Show grid
plt.legend()  # Show legend
plt.xlim(-10, 10)  # Set x-axis limits
plt.ylim(-10, 10)  # Set y-axis limits
plt.show()  # Display the plot
```

```python
# define function
def f(x):
    return x**3-2*x-5
```

```python
# imple of bissecction method

def bisection(f,x1,x2,tol=1e-6):
    if(f(x1)*f(x2)>0):
        return f"Change the choice of interval"
    
    # searching for root within 500 iterations
    for i in range(500):
        xm = (x1+x2)/2
        if(f(x1)*f(xm)<0):
            x2 = xm
        else:
            x1 = xm

        if(abs(f(xm))<tol):
            return f"root found at x = {round(xm,2)} in {i+1} iterations"
        
    return f"can't find the root"           
```

```python
# initial guesses I
x1 = -100
x2 = 258
import time as t
tStart = t.time()
print(bisection(f,x1,x2))
tEnd = t.time()
time = tEnd - tStart
print("time =",time,"sec")
```

```python
# initial guesses II
x1 = -2
x2 = 2
import time as t
tStart = t.time()
print(bisection(f,x1,x2))
tEnd = t.time()
time = tEnd - tStart
print("time =",time,"sec")
```

```python
# define function and its derivative
def f(x):
    return x**(1/3)

def df(x):
    return (1/3)*x**(-2/3)
```

```python
def newton(f, df, x0, tol):
    for i in range(500):
        x = x0 - f(x0)/df(x0)
        print(x)
        if abs(f(x)) < tol:
            return f"root found at x = {round(x,2)} in {i+1} iterations"
        x0 = x

    return f"root not found"
```

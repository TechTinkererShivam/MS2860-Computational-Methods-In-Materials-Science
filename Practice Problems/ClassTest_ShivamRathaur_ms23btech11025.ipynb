```python

# Analytical Solution is X = X0 cos(wt)
# X(t=0) is X0 = 1.0
# X(dt) = cos(1.dt)

import numpy as np
import matplotlib.pyplot as plt

k = 1    
m = 1   
dt = 0.01  
n = 1000 

time = np.linspace(0, n*dt, n)


x = np.zeros(n)
v = np.zeros(n)


x[0] = 1  
v[0] = 0 

x[1] = x[0] + v[0] * dt - 0.5 * (k/m) * x[0] * dt**2

for i in range(1, n-1):
    x[i+1] = x[i] * (2 - (k/m) * dt**2) - x[i-1]

# Graph
plt.plot(time, x, label="x")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for Argon
epsilon = 1.65e-21  # Depth of potential well in Joules
sigma = 3.4e-10     # Finite distance at which potential is zero in meters
mass = 6.63e-26     # Mass of Argon atom in kg

# Initial conditions
x0 = 4e-10          # Initial position in meters (4 √Ö)
v0 = 50             # Initial velocity in m/s (example value)
delta_t = 0.1e-12   # Time step in seconds (0.1 ps)
total_time = 0.5e-12  # Total simulation time in seconds (0.5 ps)

# Number of time steps
num_steps = int(total_time / delta_t)

# Arrays to store time, position, and velocity
time = np.linspace(0, total_time, num_steps)
x = np.zeros(num_steps)
v = np.zeros(num_steps)

# Initial values
x[0] = x0
v[0] = v0

# Define the Lennard-Jones potential and force functions
def lj_potential(r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

def lj_force(r):
    return 24 * epsilon * (2 * (sigma / r)**13 - (sigma / r)**7)

# Numerical integration using the Verlet method
for i in range(1, num_steps):
    # Calculate the force at the current position
    force = lj_force(x[i-1])
    
    # Update velocity and position
    v[i] = v[i-1] + (force / mass) * delta_t
    x[i] = x[i-1] + v[i] * delta_t

# Plotting x(t) vs t
plt.figure(figsize=(10, 6))
plt.plot(time * 1e12, x * 1e10)  # Convert time to picoseconds and position to angstroms for plotting
plt.title('Position of Argon Atom Over Time')
plt.xlabel('Time (ps)')
plt.ylabel('Position x (√Ö)')
plt.xlim(0, 0.5)  # x-axis in picoseconds
plt.ylim(4.0, 4.25)    # y-axis in angstroms
plt.grid()
plt.show()
```

Near 3.8 √Ö, the motion resembles that of an oscillator. This is because the forces acting on the atom near the equilibrium position (the minimum of the potential energy curve) provide a restoring force that pulls the atom back when it is displaced. As a result, the atom oscillates around this point, akin to the behavior of a mass-spring system in classical mechanics.

$ Ques: $ 
What is anharmonic motion?

$ Ans: $
Anharmonic motion refers to motion in which the restoring force is not directly proportional to the displacement from the equilibrium position. In contrast to harmonic motion, where the relationship between force and displacement is linear (like a spring), anharmonic systems exhibit a nonlinear relationship. This can lead to complexities such as changes in the oscillation frequency with amplitude, as observed in real molecular systems where potential energy curves deviate from a perfect parabola.

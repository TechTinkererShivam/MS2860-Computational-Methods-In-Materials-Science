```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
m = 1.0  # mass (kg)
k = 1.0  # spring constant (N/m)
dt = 0.01  # time step (s)
total_time = 10.0  # total time of simulation (s)
num_steps = int(total_time / dt)  # number of steps

# Initial conditions
x = np.zeros(num_steps)
v = np.zeros(num_steps)
x[0] = 1.0  # initial position (m)
v[0] = 0.0  # initial velocity (m/s)
a = -k * x[0] / m  # initial acceleration

# Verlet Algorithm
for i in range(1, num_steps - 1):
    # Update position using Verlet formula
    x[i + 1] = 2 * x[i] - x[i - 1] + (a * dt**2)
    # Calculate new acceleration
    a = -k * x[i + 1] / m
    # Update velocity
    v[i + 1] = (x[i + 1] - x[i - 1]) / (2 * dt)

# Calculate momentum
p = m * v  # momentum (kg*m/s)

# Create time array
t = np.arange(0, total_time, dt)

# Phase Space Diagram (p vs x)
plt.figure(figsize=(12, 6))

# Plot x(t) vs t
plt.subplot(2, 1, 1)
plt.plot(t, x, label='Position x(t)')
plt.title('Position vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()



plt.tight_layout()
plt.show()
```

```python
# Plot v(t) vs t
plt.subplot(2, 1, 2)  # Next subplot
plt.plot(t, v, color='green', label='Velocity vs Time')
plt.title('Velocity vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()

plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Phase Space Diagram (p vs x)
plt.figure(figsize=(12, 6))
plt.plot(x, p, label='Phase Space Diagram')
plt.title('Phase Space Diagram (Momentum vs Position)')
plt.xlabel('Position (m)')
plt.ylabel('Momentum (kg*m/s)')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.show()
```

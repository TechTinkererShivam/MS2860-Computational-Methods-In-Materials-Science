```python
from math import sin, exp, fabs

# code of bisection method
def bisection(f, x0, x1, tol = 1e-6):
    for i in range(500):
        x = (x0 + x1) / 2
        if fabs(f(x)) < tol:
            return f"root at {x:.5f} found in {i + 1} iterations"
        if f(x)*f(x0) < 0:
            x1 = x
        else:
            x0 = x
    return f"could not find in {i + 1} iterations, final value: {f(x)} at x = {x:.5f}"
```

```python
# defining the function 
def f(x):
    return 2 * sin(x) - exp(x)/4 - 1

# calling the function
print(bisection(f, -7, -5))
print(bisection(f, -5, -3))
```

```python
from math import sin, exp, fabs

# code of secant method
def secant(f, x0, x1, tol = 1e-6):
    for i in range(500):
        x = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        if f(x)*f(x0) < 0:
            x1 = x
        else:
            x0 = x
        if fabs(f(x)) < tol:
            return f"root at {x:.5f} found in {i + 1} iterations"
        
    return f"could not find in {i + 1} iterations, final value: {f(x)} at x = {x:.5f}"
```

```python
# defining the function
def f(x):
        return 2 * sin(x) - exp(x)/4 - 1

# calling the function
print(secant(f, -7, -5))
print(secant(f, -5, -3))
```

Ans: Secant Method took 10-12 less iterations compared to Bisection Method.

```python
# code of regula falsi method
def regula_falsi(f, x0, x1, tol = 1e-6):
    for i in range(500):
        x = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        if f(x) == -f(x0):
            x1 = x
        else:
            x0 = x
        if fabs(f(x)) < tol:
            return f"root at {x:.5f} found in {i + 1} iterations"
    return f"could not find in {i + 1} iterations, final value: {f(x)} at x = {x:.5f}"
```

```python
# defining the function
def f(x):
        return 2 * sin(x) - exp(x)/4 - 1

# calling the function
print(regula_falsi(f, -7, -5))
print(regula_falsi(f, -5, -3))
```

Ans: Regula Falsi method took less iterations than Bisection Method but more iterations than Secant Method.

Ans: The secant method generally converges faster than bisection or linear interpolation because:

* Higher Order Convergence: The secant method has a convergence rate of approximately 1.618 (the golden ratio), 
      which is faster than the linear convergence rate of bisection.

* No Interval Restriction: Unlike the bisection method, which requires the root to be within an interval, the secant 
      method uses two previous points to estimate the next point, allowing for faster movement towards the root.

* Fewer Function Evaluations: Since the secant method doesn't require recalculating the midpoint at each step, it 
      generally needs fewer iterations to meet a given stopping tolerance compared to bisection or linear interpolation.

```python
# Code of Newton's Method
def newton(f, f1, x0, tol = 1e-6):
    for i in range(500):
        x = x0 - f(x0)/f1(x0)
        if fabs(f(x)) < tol:
            return f"root at {x:.5f} found in {i + 1} iterations"
        x0 = x
    return f"could not find in {i + 1} iterations, final value: {f(x)} at x = {x:.5f}"
```

Using Newton-Raphson Formula,
$$ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} $$
$$ âx_{n+1} = x_n - \frac{x_n^2-N}{2x_n} $$
$$ âx_{n+1} = \frac{2x_n^2-x_n^2+N}{2x_n} $$
$$ âx_{n+1} = \frac{x_n^2+N}{2x_n} $$

$$ x_{n+1} = \frac{1}{2} \left(x_n+\frac{N}{x_n}\right)$$
